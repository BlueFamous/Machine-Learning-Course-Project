```python
out = []
element = []
cons = [0, 0]
sum = 0
output = 'D:\\机器学习大作业\\predict_final_new.csv'
for i in range(len(length)):
    loc = location[i]
    if(data[loc][0] == 12):
        if(len(str(data[loc][0])) == 2):
            file_path = 'D:\\机器学习大作业\\new\\0' + str(data[loc][0]) + '\\' +data[loc][1]
        else:
            file_path = 'D:\\机器学习大作业\\new\\00' + str(data[loc][0]) + '\\' +data[loc][1]
    
    for k in range(sum, sum + length[i] - 1):
        if(predict[k] == 1):
            cons[1] += 1
        else:
            cons[0] += 1
    element.append(data[loc][1])
    if(cons[0] > cons[1]):
        element.append(0)
    else:
        element.append(1)
    out.append(element)
    element = []
    sum += length[i]
    cons = [0, 0]

out = pd.DataFrame(out)
out = out.to_csv(output, index = False, header = False)
```
